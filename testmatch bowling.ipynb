```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import re
import time
```

```python
#Loading the webpage
url = 'https://stats.espncricinfo.com/ci/engine/stats/index.html?class=1;filter=advanced;orderby=wickets;template=results;type=bowling'
header = {'User Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/94.0.4606.81 Safari/537.36'}
r = requests.get(url) #fetching the url
soup = BeautifulSoup(r.content,'lxml') #getting the content
#print(soup.prettify()) #checking the data
Player_table = soup.find_all("table",class_ = 'engineTable')[2] #getting table from page
```

```python
#Creating a template to extract data from the table in webpage as per details present on page

for name in Player_table.find_all('tbody'):
    rows = name.find_all('tr')
    for row in rows:
        player = row.find('td',class_ = 'left').text
        span = row.find_all('td')[1].text
        mat = row.find_all('td')[2].text
        inns = row.find_all('td')[3].text
        balls = row.find_all('td')[4].text
        runs = row.find_all('td')[5].text
        Wkts = row.find_all('td')[6].text
        BBI = row.find_all('td')[7].text
        BBM = row.find_all('td')[8].text
        Ave = row.find_all('td')[9].text
        Econ = row.find_all('td')[10].text
        SR = row.find_all('td')[11].text
        fivewickets = row.find_all('td')[12].text
        tenwickets = row.find_all('td')[13].text
        
        
        
        #Storing it in a dictionary
        Cricketers = {'Player':player,"Span":span,"testMatches":mat,"Innings":inns,"Balls":balls,"Runs":runs,"Wkts":Wkts,
                     "BBI":BBI,"BBM":BBM,"AVG":Ave,"Economy":Econ,"Strikerate":SR,"5":fivewickets,"10":tenwickets}
       # print(Cricketers)
```

```python
#Assigning an empty list to store all dictionaries of players

Cricketers_List = []

#Automating scrapping through different pages[the code has only first 2 page loop] by using a for loop and an f string edit in the link.

for page in range(1,63):  # 62 pages thus taken 63
    #class changed to 1 as it is testmatched from website, type=bowling
    source = requests.get(f'https://stats.espncricinfo.com/ci/engine/stats/index.html?class=1;filter=advanced;orderby=wickets;page={page};template=results;type=bowling')
    
    soup = BeautifulSoup(source.content,'lxml')
    Player_table = soup.find_all("table",class_ = 'engineTable')[2]
    
    #Code from the above cell which builds table structure
    
    for name in Player_table.find_all('tbody'):
        rows = name.find_all('tr')
        for row in rows:  # using [] as per the table present on website
            player = row.find('td',class_ = 'left').text
            span = row.find_all('td')[1].text
            mat = row.find_all('td')[2].text
            inns = row.find_all('td')[3].text
            balls = row.find_all('td')[4].text
            runs = row.find_all('td')[5].text
            Wkts = row.find_all('td')[6].text
            BBI = row.find_all('td')[7].text
            BBM = row.find_all('td')[8].text
            Ave = row.find_all('td')[9].text
            Econ = row.find_all('td')[10].text
            SR = row.find_all('td')[11].text
            fivewickets = row.find_all('td')[12].text
            tenwickets = row.find_all('td')[13].text
        
        
        #Storing it in a dictionary
            Cricketers = {'Player':player,"Span":span,"testMatches":mat,"Innings":inns,"Balls":balls,"Runs":runs,"Wkts":Wkts,
                     "BBI":BBI,"BBM":BBM,"AVG":Ave,"Economy":Econ,"Strikerate":SR,"5":fivewickets,"10":tenwickets}
            Cricketers_List.append(Cricketers)
    
          #  print('Saving: ', Cricketers['Player'])

           # time.sleep(3)
```

```python
#Converting to DataFrame using pandas [sorted by number of runs by default]

df_bowling_test = pd.DataFrame(Cricketers_List)
df_bowling_test
```

```python
df_bowling_test.set_axis(['Player','Test_Span','TEST_Matches','TEST_Innings','TEST_ballsbowled','TEST_Runs_given','TEST_wickets','TEST_bestbowlinginnings','TEST_bestbowlingmatch','TEST_BOWL_AVERAGE','TEST_Economy','TEST_Bowling_Strikerate','TEST_5_wickets','TEST_10_wickets'],axis=1,inplace=True)

                            
```
